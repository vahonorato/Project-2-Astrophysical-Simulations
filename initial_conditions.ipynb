```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
```

```python
np.random.seed(0)
```

```python
N_part = 10**4
for i in range(0, N_part):
    x = np.random.uniform(0, 1, N_part)
    y = np.random.uniform(0, 1, N_part)
    z = np.random.uniform(0, 1, N_part)
```

```python
fig = plt.figure(figsize=(8,7))
ax = fig.add_subplot(projection='3d')

ax.scatter(x, y, z, edgecolor = 'black', s = 5, color = 'red')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
#for r_s fixed
r_s = 1 *u.kpc

for i in range(0, N_part):
    x1 = np.random.uniform(0, 1, N_part)
    x2 = np.random.uniform(0, 1, N_part)
    z = (1 - 2*x1)*r_s
    x = np.sqrt(r_s**2 - z**2) * np.cos(2*np.pi*x2)
    y = np.sqrt(r_s**2 - z**2) * np.sin(2*np.pi*x2)
```

```python
fig = plt.figure(figsize=(8,7))
ax = fig.add_subplot(projection='3d')

ax.scatter(x, y, z, edgecolor = 'black', s = 5, color = 'red')
ax.set_xlabel('x [kpc]')
ax.set_ylabel('y [kpc]')
ax.set_zlabel('z [kpc]')
plt.show()
```

```python
#for different r_s
r_max = 1*u.kpc
for i in range(0, N_part):
    x1 = np.random.uniform(0, 1, N_part)
    x2 = np.random.uniform(0, 1, N_part)
    x3 = np.random.uniform(0, 1, N_part)
    r_s = x3*r_max
    z = (1 - 2*x1)*r_s
    x = np.sqrt(r_s**2 - z**2) * np.cos(2*np.pi*x2)
    y = np.sqrt(r_s**2 - z**2) * np.sin(2*np.pi*x2)
```

```python
fig = plt.figure(figsize=(8,7))
ax = fig.add_subplot(projection='3d')

ax.scatter(x, y, z, edgecolor = 'black', s = 5, color = 'red')
ax.set_xlabel('x [kpc]')
ax.set_ylabel('y [kpc]')
ax.set_zlabel('z [kpc]')
plt.show()
```
